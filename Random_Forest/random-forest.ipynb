```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
df = pd.DataFrame(digits.data)
df.head()
```

```python
df['target'] = digits.target
```

```python
df.head()
```

```python
X = df.drop('target', axis= 'columns')
X.head()
```

```python
y = df['target']
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y ,test_size= 0.2, random_state = 124)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
y_predicter = model.predict(X_test)
```

# Confusion Martrix

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_predicter)
cm
```

```python
import seaborn as sns

plt.figure(figsize = (10, 7))
sns.heatmap(cm, annot = True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```

Use famous iris flower dataset from sklearn.datasets import load_iris to predict flower species using random forest classifi

```python
from sklearn.datasets import load_iris

iris = load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
X = df.drop('target', axis = 'columns')
X.head()
```

```python
y = df['target']
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state = 124)
```

```python
X_train.head()
```

```python
y_test.head()
```

```python
model = RandomForestClassifier(n_estimators= 10)
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
y_predicts = model.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_predicts)
cm
```

```python
plt.figure(figsize = (10,7))
sns.heatmap(cm, annot = True)
plt.xlabel('Predicter')
plt.ylabel('Truth')
plt.show()
```

```python
from sklearn.metrics import classification_report

cr = classification_report(y_test, y_predicts)
print(cr)
```
