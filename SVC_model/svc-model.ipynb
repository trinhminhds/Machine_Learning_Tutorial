```python
import numpy as np
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df[df.target == 0].head()
```

```python
df[df.target == 1].head()
```

```python
df[df.target == 2].head()
```

```python
df['flower_name'] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
df[50:55]
```

```python
df[100:105]
```

```python
X = df.drop(['target', 'flower_name'], axis = 'columns')
X.head()
```

```python
y = df['target']
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y , test_size= 0.2, random_state= 124)
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.svm import SVC

model = SVC()
model.fit(X_train, y_train)
```

```python
pre = model.predict(X_test)
pre[:5]
```

```python
y_test[:5]
```

```python
model.score(X_test, y_test)
```

# Ture Parameters

# 1. Regularrization (C)

```python
model_C = SVC(C = 1.0)
model_C.fit(X_train, y_train)
model_C.score(X_test,y_test)
```

```python
model_C = SVC(C = 10.0)
model_C.fit(X_train, y_train)
model_C.score(X_test,y_test)
```

# 2. Gamma 

```python
model_gamma = SVC(gamma= 10.0)
model_gamma.fit(X_train, y_train)
model_gamma.score(X_test, y_test)
```

# 3. Kernel

```python
model_kernel_linear = SVC(kernel = 'linear')
model_kernel_linear.fit(X_train, y_train)
model_kernel_linear.score(X_test, y_test)
```

```python
model_kernel_rbf = SVC(kernel = 'rbf')
model_kernel_rbf.fit(X_train, y_train)
model_kernel_rbf.score(X_test, y_test)
```

```python
model_kernel_s = SVC(kernel = 'sigmoid')
model_kernel_s.fit(X_train, y_train)
model_kernel_s.score(X_test, y_test)
```

```python
model_kernel_poly = SVC(kernel = 'poly')
model_kernel_poly.fit(X_train, y_train)
model_kernel_poly.score(X_test, y_test)
```
